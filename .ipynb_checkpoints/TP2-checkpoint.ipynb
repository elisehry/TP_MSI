## Tests statistiques
**Question 3:** La fertilité se répartit-elle suivant une loi normal en acceptant un risque de 10% ?

```r
alpha=0.10
norm_fertility = shapiro.test(donnees$Fertility)
norm_fertility
# Le test de Shapiro-Wilk est un test permettant de savoir si une série de données suit une loi normale
```

```r
if (norm_fertility$p.value>alpha){
  print("H0 accepté")
} else {
  print("H0 rejeté")
}
```

La fertilité s répartit comme une loi normale en acceptant un risque de 10%.

**Question 4:** Estimateurs de vraissemblance de la moyenne et de la variance de la variable fertilité

```r
x_barre = mean(donnees$Fertility) # Estimateur de m
s_carree = sd(donnees$Fertility) # Estimateur de sigma_carree
# Plot de la répartition des valeurs effectives prises par la variable (histogramme)
hist(donnees$Fertility,freq=FALSE,border="blue", col="lightblue", xlab="Fertilité", ylab="Densité", main="Histogramme de la fertilité", cex.axis=0.8, cex.lab=0.8)
# Plot de la courb suivie par la loie normale (lwd détermine l'épaisseur du trait)
curve(dnorm(x,x_barre,s_carree),add=TRUE,col="red",lwd=3) 
```

**Question 5:** On effectue un test sur la moyenne avec sigma inconnu -> la fonction pivot suit une loi de student à n-1 degrés de liberté : t.test

```r
alpha = 0.01
m0 = 80
test_chi_deux = t.test(donnees$Fertility, mu=m0, conf.level=1-alpha, alternative='greater')
test_chi_deux
```

```r
if (test_chi_deux$p.value>alpha){ 
  print("H0 accepté")
} else {
  print("H0 rejeté")
}
```

On ne peut pas dire que la fertilité moyenne suisse est inférieur à celle française avec un risque inférieur à 1%.

**Question 9 :** On cherche le meilleur modèle pour expliquer la fertilité

```r
RegMulti = lm(Fertility~., data = donnees)
summary(RegMulti)
```

Au regard des probabilités critiques des tests de Student et de nullité des coefficients, nous conservons toutes les variables sauf celle Examination.

Les « Coefficients » sont les valeurs que le modèle estime, leur écart-type, leur t-value et la probabilité de rejeter l’hypothèse H0 pour laquelle le coefficient est égal à 0. 

Pr(>|t|) : correspond à la probabilité critique, la valeur doit donc être faible, de préférence infèrieure à 5%.


**Question 10 :** Construction d'un nouveau modèle pour expliquer la fertilité avec 3 variables explicatives à l'aide du critère BIC qui vise la sélection de variables statistiquement significative dans le modèle. Le meilleur modèle est celui possédant le BIC le plus faible.

```r
RegMultiShort = lm(Fertility~Education+Infant.Mortality+Catholic, data = donnees)
paste('Résultat BIC (Fertility - Autres variables) :', BIC(RegMulti))
paste('Résultat BIC (Fertility - Education+Infant.Mortality+Catholic) :',BIC(RegMultiShort))
```

**Question 11:** Pour connaître le meilleur modèle qu'il est possible de construire à partir des variables explicatives disponibles, nous utilisons la fonction regsubsets de la librairie leaps qui évalue pour divers critères (BIC, R2, Cp de Mallows ...) le meilleur modèle à 1 jusqu'à nvmax variables explicatives.

```r
library(leaps)
choice = regsubsets(Fertility~., data = donnees, nbest = 1, nvmax = dim(donnees)[2]-1)
plot(choice, scale="bic") 
RegMultiBest = lm(Fertility~Education+Infant.Mortality+Catholic+Agriculture, data = donnees)
paste('Résultat BIC (Fertility - Education+Infant.Mortality+Catholic+Agriculture) :',BIC(RegMultiBest))
```

Le meilleur modèle est celui minimisant le critère BIC, c'est-à-dire ceui reprenant toutes les variables explicatives sauf Examination.

**Question 12 :** Nous voudrions prévoir la fertilité de deux communes voisines mais non recensées pour lesquelles nous ne disposons que des informations sur l'Agriculture, l'Education, le pourcentage de Catholiques et le taux de mortalité infantile. 

```r
CommuneNew = t(matrix(c(19, 28, 7, 35, 20, 17, 96, 17), ncol = 2))
CommuneNew = as.data.frame(CommuneNew)
names(CommuneNew) = c("Education", "Infant.Mortality", "Catholic", "Agriculture")
FertilityCommuneNew = predict(RegMultiBest, CommuneNew, interval = "pred")
FertilityCommuneNew 

plot(donnees$Infant.Mortality, donnees$Fertility, main = 'Evolution de la fertilité en fonction de la mortalité infantile', xlab='Mortalité infantile', ylab='Fertilité', cex.axis=0.8, cex.lab=0.8,)
points(donnees$Infant.Mortality, predict(RegMultiBest, donnees[,c("Education", "Infant.Mortality", "Catholic", "Agriculture")]), col = "blue", pch="*")
points(CommuneNew$Infant.Mortality, predict(RegMultiBest, CommuneNew), col='red')
```

Notre modèle de prédiction semble bien fonctionner. Les points bleus (fertilité prédite grace à la meilleur  régression linéaire sur les différentes communes de notre jeu de données) sont proches des ronds noirs (fertilité réelle observée dans les différentes communes). 

Ainsi nous pouvons prédire la fertilité pour notre nouvelle commune (point rouge) en fonction de la mortalité infantile, agriculture, % de catholique et l'éducation. 

**Question 13 :** Intervalles de confiance sur le taux moyen de mortalité infantile

```r
alpha95=qnorm(1-0.05/2)

eps95=alpha95*sd(donnees$Infant.Mortality)/sqrt(length(donnees$Fertility))
ICmoins95=mean(donnees$Infant.Mortality)-eps95
ICplus95=mean(donnees$Infant.Mortality)+eps95

c(ICmoins95, ICplus95)
```

```r
alpha90=qnorm(1-0.10/2)

eps90=alpha90*sd(donnees$Infant.Mortality)/sqrt(length(donnees$Fertility))
ICmoins90=mean(donnees$Infant.Mortality)-eps90
ICplus90=mean(donnees$Infant.Mortality)+eps90

c(ICmoins90, ICplus90)
```

```r
paste('Actuellement lerreur standard est différente de 3, sd =', sd(donnees$Infant.Mortality))
```

```r
eps95_SD3=alpha95*3/sqrt(length(donnees$Fertility))
ICmoins95_SD3=mean(donnees$Infant.Mortality)-eps95_SD3
ICplus95_SD3=mean(donnees$Infant.Mortality)+eps95_SD3

c(ICmoins95_SD3, ICplus95_SD3)
```
